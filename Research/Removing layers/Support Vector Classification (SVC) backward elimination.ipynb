## Functions

```python
import pandas as pd

def Select_The_Features(features, true_or_false):
  selected_features = []

  for idx, selection in enumerate(true_or_false):
    if selection == True:
      selected_features.append(features[idx])

  return selected_features
```

```python
import pandas as pd

def Connect_Feature_Names_And_Data(names, data):
    # Create an empty DataFrame
    df = pd.DataFrame()

    # Assign column names from the names array to each column in the DataFrame
    for i, name in enumerate(names):
        df[name] = data[:, i]

    return df
```

## Select features

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

n_features_to_select = 6

estimator = SVC(kernel="linear")
selector = RFE(estimator, n_features_to_select=n_features_to_select, step=1)
selector = selector.fit(X, y)
```
